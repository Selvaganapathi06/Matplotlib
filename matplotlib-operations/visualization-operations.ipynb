```python
#Assigning values x,y axis
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
x =[1,2,3,4,5,6]
y =[7,8,9,10,11,12]
plt.title("Textile sales data")
plt.xlabel("Mens Dresses")
plt.ylabel("Womens Dress")
plt.plot(x,y)
plt.show()
```

```python
#comparing values x,y,z axis
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
x =[1,2,3,4,5,6]
y =[7,8,9,10,11,12]
z =[13,14,15,16,17,18]
plt.title("Textile sales data")
plt.xlabel("Mens Dresses")
plt.ylabel("Womens Dress")
plt.plot(x,y)
plt.plot(x,z)
plt.legend(['THis is x,y','This is x,z'])
plt.show()
```
